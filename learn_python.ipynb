```python
# compute the x and y coordinate for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up a subplot grid that has height 2 and width 1
plt.subplot(2,1,1)

#make a 1st plot
plt.plot(x,y_sin,'r--')
plt.title("sine")

# set 2nd plot
plt.subplot(2,1,2)
#make a 2nd plot
plt.plot(x,y_cos,"b*")
plt.title("cosine")

# show the figure
plt.show()
```

# piechart

```python
# Data to plot 
labels='python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0) # (1st slice,2nd slice,3rd slice,4th slice) explode  1st slice ( how much far 1st slice go)

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,   
autopct='%1.1f%%' ,shadow=True) # format of values , shadow will create shadow

plt.axis('equal')
plt.show()
```

# 


DATA CLEANING
### We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the passengers(imputation. However we can be smarter about this and check the average age by passenger class

```python

# before removing age column we are relationship between passenger class and age( with respect to passenger class we are going to find what is the average of the people)
# using boxplot()
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
# based on this passengerclass and age we are going to replace NAN value in the age column
def impute_age(cols):
    Age=Age = cols.iloc[0]
    Pclass=Age = cols.iloc[1]
    if pd.isnull(Age):
       if Pclass==1:
             return 37
       elif Pclass==2:
           return 29
       else:
           return 24
    else:
        return Age
    
```

```python
# Now apply this function
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1) # .apply() is going to apply impute_age function on each and every record on passenger class and age
```

```python
# now again checking weather the null values is replaced or not after performing imputation
sns.heatmap(train.isnull(),yticklabels=False,cmap='viridis')
```

### we can see age column is not showing NAN becuse it is replaced using imput_age function but in cabin we can it contain many null values so to handle this require feature engeneering for now we are not performing feature engineering so we are going to drop cabin column

```python
train.drop('Cabin',axis=1,inplace=True)
```

# Building Logistic Regression model
### lets start by splitting our data into a training set and test set

# Train Test Split

```python
train.drop('Survived',axis=1).head()
```

```python
train['Survived'].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(train.drop('Survived',axis=1),
                                               train['Survived'],test_size=0.30,
                                               random_state=101)
```

# Training and predicting

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(solver='liblinear', max_iter=1000)
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,predictions)
accuracy
```
